```python
# csv 파일 받아오기
import pandas as pd
df = pd.read_csv('csv/전국초중등학교위치표준데이터.csv', encoding='CP949')

df.head()
```

```python
df2 = df[df['학교명'] == '대정고등학교']

df2
```

```python
###### 고등학교 명단과 필요한 데이터 추출하기 ######

df2 = df[df['학교급구분'] == '고등학교']

high_schools_name = list(set(df2.loc[:, '학교명'].to_numpy()))

high_schools_name.remove('목도고등학교') # 폐교
high_schools_name.remove('이설광주예술고등학교') # 데이터 불충분

high_schools_name
```

```python
# 네이버에 검색한 고등학교 데이터 크롤링
import requests 
from bs4 import BeautifulSoup

def get_high_school_information(high_school_name: str):
    url = url_information = f'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query={high_school_name}'

    response = requests.get(url)

    html = response.text

    soup = BeautifulSoup( html , 'html.parser' )

    try:
        a_tag       = soup.find('a', class_ = 'home')
        information = soup.find('dl', class_ = 'info').find_all('dd')
    except:
        print(high_school_name)
    
    # 리스트 타입으로 바꿈
    list_information = []
    for i in range(len(information)):
        list_information.append(information[i].text)

    # 받아온 리스트 중에 '방과후학교'가 있으면 제거함(불필요)
    for i in list_information:
        if '방과후학교' in i:
            list_information.remove(i)

    # 홈페이지 주소
    link = a_tag.get('href')
    
    list_information.append(link)

    return list_information

```

```python
# 시험
get_high_school_information('대정고등학교')
```

```python
import json

# 위도 경도 함수 ---- kako api
def get_coordinate(addr):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={address}'.format(address=addr)
    headers = {"Authorization": "KakaoAK " + 'cba6eb29be7c48e97d9a5cd246429cdd'}
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = result['documents'][0]['address']
    
    return float(match_first['x']), float(match_first['y'])
```

```python
#시험
get_coordinate('울산과학고')
```

```python
# 고등학교 이름 넣으면 데이터들을 크롤링하여 리스트 형태로 변환하여 딕셔너리에 넣음
dict_high_schools = {} # 데이터를 넣을 딕셔너리

def put_high_school_information_in_dict(high_shcool_name: str):
    # 인덱스 번호
    # 0 : 주소
    # 1 : 전화번호
    # 3 : 학생수
    # 5 : 홈페이지
    high_school_information = get_high_school_information(high_shcool_name)

    address  = high_school_information[0]
    # 전화번호에 공백에 있는 경우를 대비하여 공백을 제거함
    phone    = high_school_information[1].replace(" ", "")
    
    split_list_student = high_school_information[3].split(' ')
    students = split_list_student[0]

    # 문자열 '남' 다음 인덱스가 남학생 수
    boy  = split_list_student[split_list_student.index('남') + 1]
    girl = split_list_student[split_list_student.index('여') + 1]

    link = high_school_information[5]

    try:
        longitude, latitude = get_coordinate(address)
    except:
        print(high_shcool_name)

    # 딕셔너리에 데이터 넣기
    dict_high_schools[high_shcool_name] = {
        '위도' : latitude,
        '경도' : longitude,
        '주소' : address,
        '전화번호' : phone,
        '재학생 수' : students,
        '남학생 수' : boy,
        '여학생 수' : girl,
        '홈페이지' : link
    }
```

```python
# 시험
put_high_school_information_in_dict('속초고')

dict_high_schools
```

```python
# 전부 집어넣기
for i in high_schools_name:
    put_high_school_information_in_dict(i)

high_schools_dataFrame = pd.DataFrame(dict_high_schools).T
high_schools_dataFrame
```
